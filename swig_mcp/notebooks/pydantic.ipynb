```python
from typing import List   
```

```python
class user(BaseModel):
    name: str
    age: int
    email: str
    friends: List[str] = Field(default_factory=list)
```

# Play with Pydantic Custome Validator

```python
from pydantic import BaseModel, EmailStr, validator


class User(BaseModel):
    username: str
    age: int
    email: EmailStr
```

```python
validator("username")
def no_spaces(cls, value):
        if " " in value:
            raise ValueError("username must not contain spaces")
        return value
```

```python
user = User(
    username="john_doe.   " )
```

```python
@validator("age")
def age_must_be_adult(cls, value):
        if value < 18:
            raise ValueError("age must be 18 or above")
        return value

@validator("email")
def company_email_only(cls, value):
        if not value.endswith("@company.com"):
            raise ValueError("email must be a @company.com address")
        return value
```
